B4: in ra hình tam giác có độ cao h được nhập từ bàn phím

```python
h = int(input())
for i in range(1, h+1):
    for j in range(1, 2*h):
        if i == h or j == h - i + 1 or j == h + i - 1:
            print("*", end="")
        else:
            print(" ", end="")
    print()
```

B19: dãy fibonacci: 1,1,2,3,5,8,13,....A nguyên dương, tìm số trong dãy fibonacci lớn nhất <= A

```python
A = int(input())
f1, f2 = 1, 1
fm = 1
while f2 <= A:
    fm = f2
    f1, f2 = f2, f2 + f1
print(fm)
```

A6: nhập a, b. tính tổng các số nguyên tố trong [a,b]

```python
def kiem_tra(n):
    if n < 2:
        return False
    for i in range(2, int(n**0.5)+1):
        if n%i == 0:
            return False
    return True
a,b = map(int, input().split())
tong = 0
for i in range(a,b+1):
    if kiem_tra(i):
        tong += i
print(tong)
```
