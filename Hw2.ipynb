B3: nhập vào số a, b, c, kiểm tra xem có phải tam giác không, kiểm tra đều, cân, vuông, thường. Tính diện tích, chu vi

```python
a, b, c = map(float, input().split())
if a<= 0 or b <= 0 or c <= 0:
    print("giởn mặt hả???")
elif a+b>c and a+c>b and c+b>a:
    p=(a+b+c)/2
    s=(p*(p-a)*(p-b)*(p-c))**0.5
    if a==b==c:
        print(f"{s} \n{p*2} \nĐều")
    elif (a==b!=c or a==c!=b or b==c!=c) and (a**2+b**2==c**2 or a**2+c**2==b**2 or b**2+c**2==a**2):
        print(f"{s} \n{p*2} \nVuông cân")
    elif (a==b!=c or a==c!=b or b==c!=c) and (a**2+b**2!=c**2 or a**2+c**2!=b**2 or b**2+c**2!=a**2):
        print(f"{s} \n{p*2} \nCân")
    elif (a!=b!=c or a!=c!=b or b!=c!=c) and (a**2+b**2==c**2 or a**2+c**2==b**2 or b**2+c**2==a**2):
        print(f"{s} \n{p*2} \nVuông")
    else:
        print(f"{s} \n{p*2} \nThường")
else:
    print("không phải tam giác!!!")
```
