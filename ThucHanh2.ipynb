6: hoa => thuong va nguoc lai

```python
a = input()
a = ord(a)
if 65 <= a <= 90:
    a += 32
    print(f"{chr(a)}")
elif 97 <= a <= 122:
    a -= 32
    print(f"{chr(a)}")
else: 
    print("ko hop le")
```
