ord(c) : ordinary : mã số của c
chr

```python
c = input("Nhập vào 1 kí tự: ")
if 0<len(c)<=1:
    ma = ord(c)
    if 65<=ma<=90:
        print(chr(ma+32))
    elif 97<=ma<=122:
        print(chr(ma-32))
    else:
        print("Kí tự nhập vào không phải là 1 chữ cái!")
else: print("Bảo nhập 1 kí tự thôi mà [-_-']")        
```

Bài 6: nhập 3 số, kiểm tra là cạnh tam giác hay không, nếu là tam giác thì tính diện tích (1 chữ số thập phân).

```python
a, b, c = map(float, input("Nhập vào 3 cạnh của tam giác:").split())
p = (a+b+c)/2
S = (p*(p-a)*(p-b)*(p-c))**0.5
if 0<a and b>0 and 0<c and a+b>c and a+c>b and b+c>a:
    print("Diện tích tam giác là: " + str(round(S,1)))
else:
    print("Không phải 3 cạnh của tam giác")
```

9: Chuyển đổi số thành chữ số: Nhập một số nguyên từ 0 đến 9, in ra tên chữ số
tương ứng (0 -> "Không", 1 -> "Một", ...) – Xem trước match …case 

```python
x = int(input("Nhập vào một số nguyên(0-9): "))
match x:
    case 0: print("Không")
    case 1: print("Một")
    case 2: print("Hai")
    case 3: print("Ba")
    case 4: print("Bốn")
    case 5: print("Năm")
    case 6: print("Sáu")
    case 7: print("Bảy")
    case 8: print("Tám")
    case 9: print("Chín")
    case _: print("0-9 thoii mờ!!")
```
